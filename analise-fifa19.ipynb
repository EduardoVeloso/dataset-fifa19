```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('data.csv')
```

```python
df = data[['Name','Age','Overall','Club','Wage','Preferred Foot','International Reputation','Skill Moves','Potential','Value','Position','Jersey Number']]
```

```python
df.index = data['ID']
```

```python
df.info()
```

```python
df.head()
```

```python
#df['Wage'] = df['Wage'].fillna(0) substituir valores NaN
```

```python
df['Wage'] = df['Wage'].apply(lambda x: x[1:])
df['Wage']
```

```python
df['Wage'] = df['Wage'].apply(lambda x: x[:-1])
df['Wage']
```

```python
df['Wage'] = pd.to_numeric(df['Wage'])
```

```python
df[df['Wage'].isnull()] = 0
```

```python
df[df['Jersey Number'].isnull()] = 0
```

```python
df.info()
```

```python
sns.distplot(df['Wage'])
```

```python
sns.jointplot(x = 'Jersey Number', y = 'Wage', data = df, kind = 'scatter')
```
